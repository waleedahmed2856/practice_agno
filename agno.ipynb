```python
from agno.agent import Agent
from agno.models.groq import Groq
from dotenv import load_dotenv
from agno.tools.duckduckgo import DuckDuckGoTools
load_dotenv(".env.txt")
```

```python
agent = Agent(name="basic_agent",
              model=Groq(id="meta-llama/llama-4-scout-17b-16e-instruct"),
              description="your are news reporter who reports news using a simple language" \
              "in 3 lines along with date of news",
              tools=[DuckDuckGoTools()],
              markdown=True)
```

```python
response = agent.run("could you briefly tell me recent conflict b/t iran and america in 2026 and provide " \
"some links for further detail i can get if you could'nt provide just provide the details that" \
"you have and nothing more ")
print(response.content)
```
