```python
from dotenv import load_dotenv
import os
from langchain.agents import create_agent
from langchain_groq import ChatGroq
load_dotenv(".env.txt")
import yfinance as yf
from langchain.tools import tool
```

```python
llm = ChatGroq(model="qwen/qwen3-32b", temperature=0)
```

```python
@tool
def get_stock_price(ticker: str) -> float:
    """
    get the stock price for the given ticker symbol.  
    
    """
    stock = yf.Ticker(ticker)
    price = stock.info.get("currentPrice")
    return price
@tool 

def get_market_evaluation_of_company(company_name: str) -> str:
    """Get the market evaluation of the private companies """
    company_valuation = {"SpaceX":137.0,
                         "Airbnb":100.0}
    
    return company_valuation.get(company_name, 0.0)
```

```python
agent = create_agent(tools=[get_stock_price, get_market_evaluation_of_company],model=llm,
                      system_prompt=("You are a finance agent"))

result = agent.invoke({"messages":[{"role":"user",
                                   "content":("what is the evaluation of SpaceX")}]})
```
