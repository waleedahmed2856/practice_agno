```python
from dotenv import load_dotenv
from agno.agent import Agent
from agno.models.google import Gemini 
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.team import Team
load_dotenv(".env.txt")
```

```python
class TechnicalSupportTeam:
    __name__ = "TechnicalSupportTeam"
    
    def __init__(self, query: str) -> str:
        self.responses = {
            "app crash": "Try restarting the app. If the issue persists, clear cache or reinstall the application.",
            "login issue": "Check your credentials and internet connection. If forgotten, use the password reset option.",
            "freezing": "Close unnecessary background tasks and restart the system. Make sure your software is updated."
        }

        for key, value in self.responses.items():
            if key in query.lower():
                return value

        return "Please provide more details about technical support."
    
class SalesTool:
    __name__ = "SalesTool"
    
    def __init__(self, query: str) -> str:
        self.responses = {
            "cost": "Our premium product costs $44 and includes 24/7 support service.",
            "discount": "Yes, we currently have a 20% discount available on all products.",
            "pricing": "We offer Basic, Pro, and Premium plans starting from $9.9 per month."
        }

        for key, value in self.responses.items():
            if key in query.lower():
                return value

        return "Please provide more details about sales or pricing."

class GeneralSupportTool:
    __name__ = "GeneralSupportTool". 
    def __init__(self, query: str) -> str:
        self.responses = {
            "hours": "Our working hours are 9 AM to 5 PM.",
            "location": "Our office is located in Lahore Gulberg.",
            "contact": "You can contact us at info@gmail.com."
        }

        for key, value in self.responses.items():
            if key in query.lower():
                return value

        return "Please provide more details about general support."
    
tech_agent = Agent(name="Tech support Agent",
                       role="",
                       model=Gemini(id="gemini-2.0-flash"),
                       tools=[TechnicalSupportTeam()],
                       instructions="Use the techninal support tool to answer the user query" \
                       "provide them direct answer using information.",
                       debug_mode=False,
                       markdown=True)
    
sales_agent = Agent(name="Sales Agent",
                       role="",
                       model=Gemini(id="gemini-2.0-flash"),
                       tools=[SalesTool()],
                       instructions="Use the sales Tools and answer the pricing and discount related " \
                       "questions. provide them direct answer using information. ",
                       debug_mode=False,
                       markdown=True)
    
general_agent = Agent(name="Tech support Agent",
                       role="Answer the general questions like hours, location, contact info",
                       model=Gemini(id="gemini-2.0-flash"),
                       tools=[GeneralSupportTool()],
                       instructions="use the general info tool to help with common inquiries" \
                       "provide them direct answer using information. ",
                       debug_mode=False,
                       markdown=True)

route_team = Team(name="Customer care chatbot agent",
                  mode="route",
                  members=[tech_agent, general_agent, sales_agent],
                  model=Gemini(id="gemini-2.0-flash"),
                  instructions="Route the query to correct agent based on weather its technical,"
                  "sales, or general.",
                  markdown=True,
                  debug_mode=False)
```
