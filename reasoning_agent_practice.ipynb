```python
from dotenv import load_dotenv
from langchain.agents import create_agent
from langchain_groq import ChatGroq
load_dotenv(".env.txt")
```

```python
llm = ChatGroq(model="qwen/qwen3-32b", temperature=0)

agent = create_agent(tools=[],model=llm,
                      system_prompt=("you are a advanced reasoning assitant."
                      "List out all the steps you carry to reason with number"
                      "if you are using formula, it should not be in Latex but plain formulas"))
```

```python
result = agent.invoke({"message":[{"role":"user",
                                   "content":("how much time will be taken to travel from Isalambad "
                                   "to Lahore if I use car")}]})
```
